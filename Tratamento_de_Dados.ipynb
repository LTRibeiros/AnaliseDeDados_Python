# Tratamento dos dados

```python
dados1 = dados.drop(columns=['Carimbo de data/hora'])
dados1.head()
```

```python
dados1 = dados1.rename(columns={'Qual a sua altura?' : 'Altura'})
```

```python
dados1 = dados1.rename(columns={'Qual a sua cor favorita?' : 'Cor favorita'})
```

```python
dados1 = dados1.rename(columns={'Qual o número de seu calçado?' : 'Numero do calçado'})
```

```python
dados1
```

```python
dados1.isnull().sum()
#checa se tem algum valor vazio
```

```python
dados1.dtypes
```

```python
dados1['Altura'] = dados1['Altura'].apply(lambda x: x.replace(',','.'))
dados1
```

```python
dados1['Altura'] = dados1['Altura'].astype(float)
dados1.dtypes
```

```python
import plotly.express as px
bolhas = px.scatter(dados1, y="Altura", x="Numero do calçado")
bolhas.show()
```

```python
dados1.to_csv('dados_tratados.csv', sep=',', encoding='UTF-8', index=False)
```
