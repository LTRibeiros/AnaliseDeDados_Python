```python
dados = pd.read_csv('/content/drive/MyDrive/dados_tratados.csv', sep="," , encoding='UTF-8')
dados.head(5)
```

```python
contagem_cores = dados['Cor favorita'].value_counts() #ocorrencias de cada cor favorita
#criando um gráfico de pizza
plt.pie(contagem_cores, labels=contagem_cores.index, autopct='%1.1f%%')
plt.title('Distribuição de Cores favoritas')
plt.axis('equal')
plt.show()
```

```python
#histograma de altura
plt.hist(dados['Altura'], bins=10, edgecolor='black')
plt.xlabel('Altura')
plt.ylabel('Frequência')
plt.title('Histograma de Altura')
plt.show()
```
