```python
dados = dados.rename(columns={'ns2': 'ns10'})
dados.head()
```

```python
dados1 = dados.drop(columns=['ns1','ns10'])
dados1.head(2)
```

```python
# Exclui a primeira linha
dados2 = dados1.drop(0)
dados2.head(2)
```

Análise dos tipos de atributos:

object: strings

float64: reais

complex: complexos

```python
dados1.dtypes
```

```python
# Convertendo os valores para string e preenchendo zeros à esquerda
dados1['datas'] = dados1['datas'].astype(str).str.zfill(8)
# Convertendo para o formato de data
dados1['datas'] = pd.to_datetime(dados1['datas'], format='%d%m%Y')
# Formatando para exibir como dd/mm/aaaa
dados1['datas'] = dados1['datas'].dt.strftime('%d/%m/%Y')
```

```python
dados1
```

```python
dados1['compra'] = dados1['compra'].apply(lambda x: x.replace(',','.'))
dados1.head()
```

```python
dados1['venda'] = dados1['venda'].apply(lambda x: x.replace(',','.'))
dados1.head()
```

```python
dados1.dtypes
```

```python
dados1['compra'] = dados1['compra'].astype(float)
dados1['venda'] = dados1['venda'].astype(float)
```

```python
dados1.dtypes
```

```python
# Alterar todos de uma vez
dados1[['compra', 'venda']] = dados1[['compra', 'venda']].apply(pd.to_numeric)
```

```python
# Salvando arquivo
dados1.to_csv('dados_tratados_dolar.csv',
              encoding = 'iso-8859-1', index = False)
```
